### Newton-Raphson Visualisation

```python
import plotly.graph_objects as go
import plotly.express as px
import plotly.io as pio
pio.renderers.default = 'notebook_connected'

import numpy as np
```

```python
def f(x):
    return x**2 + 2*x - 8

def f_prime(x):
    return 2*x+2

start_x = -7
end_x = 10.5
x_temp = 10
```

```python

x_vals = np.linspace(start_x,end_x,500)
y_vals = f(x_vals)

nr_guesses = []

for i in range(3):
    nr_guesses.append((x_temp, f(x_temp), f_prime(x_temp), x_temp-f(x_temp)/f_prime(x_temp)))
    x_temp = x_temp-f(x_temp)/f_prime(x_temp)

fig = go.Figure()
fig.add_trace(go.Scatter(x=x_vals, y = y_vals, name='f(x)'))



for i, (x_temp, f_x, f_prime_x, next_guess) in enumerate(nr_guesses):
    print(f'Guess {i}: {round(x_temp,3)}')
    fig.add_trace(go.Scatter(x=[x_temp], y=[f_x], name=f'Guess {i}',
                                line_color=px.colors.qualitative.Plotly[i+1]))
    fig.add_trace(go.Scatter(   x=x_vals, 
                                y=x_vals*f_prime_x + (f_x-x_temp*f_prime_x),
                                 name=f'Tangent {i}',
                                 line_color=px.colors.qualitative.Plotly[i+1],
                                 line=dict(dash='dashdot'),
                                 opacity=0.6))
    print(f_x, f_prime_x, x_temp)

fig.update_layout(template='plotly_white',
                    xaxis=dict(title="x"),
                    yaxis=dict(title="f(x)"))
fig.update_xaxes(showgrid=False)
fig.update_yaxes(showgrid=False)
fig.update_layout(height=500, width=700)
fig.show()
```
